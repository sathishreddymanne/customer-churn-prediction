```python
y=df[['Churn']]
x=df[['Age','Gender','Tenure','MonthlyCharges']]
```

```python
x['Gender']=x['Gender'].apply(lambda x : 1 if x=='Female' else 0)
# 1 = Female  and 0 = Male 
```

```python
type(x['Gender'][0])
```

```python
y['Churn']=y['Churn'].apply(lambda x : 1 if x =='Yes' else 0 )
# 1 = yes and 0 = no
```

```python
type(y['Churn'][0])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
X_train=scalar.fit_transform(X_train)
```

```python
import joblib
joblib
```

```python
X_test=scalar.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print('Accuraacy score on model {}'.format(accuracy_score(y_test,predictions)))
```

# Linear Regrission

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
log_model =LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
log_model.predict([[34,0,10,100]])
```

```python
y1_pred=log_model.predict(X_test)
```

```python
y_test
```

```python
modelperformance(y1_pred)
```

### accuracy = 0.875

# K Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_grid={
    "n_neighbors":[3,5,7,9],
    "weights":["uniform","distance"],
}
```

```python
gridkn=GridSearchCV(KNeighborsClassifier(),param_grid,cv= 5)
```

```python
gridkn.fit(X_train,y_train)
```

```python
gridkn.best_params_
```

```python
y2_pred=gridkn.predict(X_test)
```

```python
y2_pred
```

```python
modelperformance(y2_pred)
```

### accuracy = 0.85

# Support Vector classifier

```python
from sklearn.svm import SVC
```

```python
svm=SVC()
```

```python
param_grid={
    "C":[0.01,0.1,0.5,1],
    "kernel":["linear","rbf","poly"]
}
```

```python
gridsvc=GridSearchCV(svm,param_grid,cv=5)
```

```python
gridsvc.fit(X_train,y_train)
```

```python
gridsvc.best_params_
```

```python
y3_pred=gridsvc.predict(X_test)
y3_pred
```

```python
modelperformance(y3_pred)
```

# Accuracy = 0.87

# Decision Tree


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid={
    "criterion":["gini","entropy"],
     "splitter":["best","random"],
     "max_depth":[None,10,20],
     "min_samples_split":[2,5,10],
     "min_samples_leaf":[1,2,4]
}
```

```python
grid_tree= GridSearchCV(DecisionTreeClassifier(),param_grid,cv=5)
grid_tree.fit(X_train,y_train)
```

```python
grid_tree.best_params_
```

```python
y4_pred =grid_tree.predict(X_test)
```

```python
y4_pred
```

```python
modelperformance(y4_pred)
```
